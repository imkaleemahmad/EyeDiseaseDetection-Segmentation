# Load an Image for Segmentation and Processing

# Initialize Segment Anything Model:
> Pay attention to the **pred_iou_thresh**, the lower it is, the more masks it will generate. Make it higher to generate masks only of good quality and less number of them.

> In our case we will keep it between 0.7 to 1.0 as we will be dealing with eyes. The lower the threshold the more masks it will generate, It might generate mask for veins, shadows, light reflection off the sclera if the threshold is lower.

```python
sam_checkpoint = "sam_vit_h_4b8939.pth"
model_type = "vit_h"

device = "cuda"

sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
sam.to(device=device)

### Pay attention to the pred_iou_thresh, the lower the more masks it will generate. Make it higher to generate masks only of
### good quality and less number of them
# Create a mask generator using the SamAutomaticMaskGenerator class.
mask_generator = SamAutomaticMaskGenerator(

    # Specify the model to be used for mask generation.
    model=sam,

    # Set the number of points per side for the mask generation process.
    # This determines the granularity of the generated mask.
    points_per_side=32,

    # Set a threshold for filtering the predicted masks based on their
    # Intersection over Union (IoU) score. Masks with an IoU score above
    # this threshold will be considered of high quality.
    pred_iou_thresh=0.95,

    # Set a stability score threshold. This threshold likely relates to
    # the stability of the generated masks or their quality in some way.
    stability_score_thresh=0.92,

    # Specify the number of layers to crop during mask generation.
    # This might control how many layers of the input are considered
    # when generating the mask.
    crop_n_layers=1,

    # Set a downscale factor for the number of points used in the mask
    # generation process. This factor might reduce the number of points
    # used in the process, possibly for efficiency reasons.
    crop_n_points_downscale_factor=2,

    # Set a minimum mask region area. Masks with regions smaller than
    # this area will be filtered out. This could be used to remove
    # very small or insignificant mask regions.
    min_mask_region_area=100  # Requires OpenCV to run post-processing
)
```

# Declare function to visualize the masks generated together with a number that identifies each of them:
> We can use each one of the detected generated segmentation masks for further processing.



```python
def show_anns(anns):
    if len(anns) == 0:
        return

    # Sort masks by area in descending order
    sorted_anns = sorted(enumerate(anns), key=(lambda x: x[1]['area']), reverse=True)
    ax = plt.gca()

    # Disable autoscale to keep the image size consistent
    ax.set_autoscale_on(False)

    # Iterate through each mask and display it on top of the original image
    for original_idx, ann in sorted_anns:
        m = ann['segmentation']
        img = np.ones((m.shape[0], m.shape[1], 3))

        # Generate a random color for the mask
        color_mask = np.random.random((1, 3)).tolist()[0]
        for i in range(3):
            img[:,:,i] = color_mask[i]

        # Blend the mask with the image, using 0.35 as the alpha value for transparency
        ax.imshow(np.dstack((img, m*0.35)))

        # Find contours of the mask to compute the centroid
        contours, _ = cv2.findContours(m.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if contours:
            cnt = contours[0]
            M = cv2.moments(cnt)

            # Compute the centroid of the mask if the moment is non-zero
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])

                # Display the original index number (1-based) at the centroid of the mask
                # The text is white, bold, and has a font size of 16
                ax.text(cx, cy, str(original_idx), color='white', fontsize=16, ha='center', va='center', fontweight='bold')

```

## Displaying multiple segmented objects/parts with index numbers:
> In this cell, all the indexes are stored in a list and are displayed combined in one image.

> It then assigns the unique colors to the segmented mask indices and overlaying these colored masks onto a blank canvas, creating a visualization of segmented regions.

```python
# Assuming masks is a list of dictionaries, each containing a 'segmentation' mask
mask_index = [0, 2, 4]

# Dynamically generate unique RGB colors for each index using a colormap
num_colors = len(mask_index)
colormap = plt.cm.jet  # You can change this to any other available colormap
colors = {index: (np.array(colormap(i / (num_colors - 1))[:3]) * 255).astype(int) for i, index in enumerate(mask_index)}

# Initialize an RGB image with the same shape as one of the segmentation masks but with an additional dimension for color
height, width = masks[0]['segmentation'].shape
combined_image = np.zeros((height, width, 3), dtype=np.uint8)

# Apply color and combine
for index in mask_index:
    mask = masks[index]['segmentation']
    color = colors[index]
    for c in range(3):  # Iterate over RGB channels
        # Intermediate computation in higher data type and then clipping
        channel_data = (combined_image[..., c].astype(np.int32) + (mask * color[c])).clip(0, 255)
        combined_image[..., c] = channel_data.astype(np.uint8)

# Convert to a format that can be displayed
stable_diffusion_image = PIL.Image.fromarray(combined_image)
display(stable_diffusion_image)
```
